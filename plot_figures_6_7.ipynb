```python
# bin the data and calculate the KL divergence
# first, let's bin the data
bin_size_on_SNR = 0.50
SNR_bins = np.arange(1, 5.01, bin_size_on_SNR)
value_bins = np.arange(0, 1.01, 0.01)
value_bins_for_SSmag = np.arange(0, 51, 1)
# for every bin, find the data corresponding to that bin, than calculate the histogram, then calculate the KL divergence
KL_divergence_R_value_raw = []
KL_divergence_CIPSR_raw = []
KL_divergence_Schuster_raw = []
KL_divergence_DL_raw = []

KL_divergence_R_value_bandpass = []
KL_divergence_CIPSR_bandpass = []
KL_divergence_Schuster_bandpass = []
KL_divergence_DL_bandpass = []

KL_divergence_R_value_PAC = []
KL_divergence_CIPSR_PAC = []
KL_divergence_Schuster_PAC = []
KL_divergence_DL_PAC = []

# a small number to avoid division by zero
epsilon = 1e-15

for SNR_bin in SNR_bins:
    # find the data corresponding to this bin
    indices_sine = np.where((Peak_SNR_of_raw_sin_wave_peak >= SNR_bin) & (Peak_SNR_of_raw_sin_wave_peak < SNR_bin + bin_size_on_SNR))[0]
    indices_noise = np.where((Peak_SNR_of_raw_noise_peaks >= SNR_bin) & (Peak_SNR_of_raw_noise_peaks < SNR_bin + bin_size_on_SNR))[0]

    # skip if there is no data in this bin
    if len(indices_sine) == 0 or len(indices_noise) == 0:
        continue

    # calculate the histogram
    hist_R_value_raw_sine, _ = np.histogram(np.array(R2_value_of_raw_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_R_value_raw_noise, _ = np.histogram(np.array(R2_value_of_raw_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_CIPSR_raw_sine, _ = np.histogram(np.array(CIPSR_of_raw_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_CIPSR_raw_noise, _ = np.histogram(np.array(CIPSR_of_raw_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_Schuster_raw_sine, _ = np.histogram(np.array(Schuster_mag_of_raw_sin_wave_peak)[indices_sine], bins=value_bins_for_SSmag, density=True)
    hist_Schuster_raw_noise, _ = np.histogram(np.array(Schuster_mag_of_raw_noise_peaks)[indices_noise], bins=value_bins_for_SSmag, density=True)
    hist_DL_raw_sine, _ = np.histogram(np.array(DL_prob_of_raw_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_DL_raw_noise, _ = np.histogram(np.array(DL_prob_of_raw_noise_peaks)[indices_noise], bins=value_bins, density=True)
    
    # calculate the KL divergence
    KL_divergence_R_value_raw.append(entropy(hist_R_value_raw_sine + epsilon, hist_R_value_raw_noise + epsilon))
    KL_divergence_CIPSR_raw.append(entropy(hist_CIPSR_raw_sine + epsilon, hist_CIPSR_raw_noise + epsilon))
    KL_divergence_Schuster_raw.append(entropy(hist_Schuster_raw_sine + epsilon, hist_Schuster_raw_noise + epsilon))
    KL_divergence_DL_raw.append(entropy(hist_DL_raw_sine + epsilon, hist_DL_raw_noise + epsilon))

    # now do the same for bandpass filtered data
    hist_R_value_bandpass_sine, _ = np.histogram(np.array(R2_value_of_bandpass_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_R_value_bandpass_noise, _ = np.histogram(np.array(R2_value_of_bandpass_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_CIPSR_bandpass_sine, _ = np.histogram(np.array(CIPSR_of_bandpass_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_CIPSR_bandpass_noise, _ = np.histogram(np.array(CIPSR_of_bandpass_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_Schuster_bandpass_sine, _ = np.histogram(np.array(Schuster_mag_of_bandpass_sin_wave_peak)[indices_sine], bins=value_bins_for_SSmag, density=True)
    hist_Schuster_bandpass_noise, _ = np.histogram(np.array(Schuster_mag_of_bandpass_noise_peaks)[indices_noise], bins=value_bins_for_SSmag, density=True)
    hist_DL_bandpass_sine, _ = np.histogram(np.array(DL_prob_of_bandpass_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_DL_bandpass_noise, _ = np.histogram(np.array(DL_prob_of_bandpass_noise_peaks)[indices_noise], bins=value_bins, density=True)

    # calculate the KL divergence
    KL_divergence_R_value_bandpass.append(entropy(hist_R_value_bandpass_sine + epsilon, hist_R_value_bandpass_noise + epsilon))
    KL_divergence_CIPSR_bandpass.append(entropy(hist_CIPSR_bandpass_sine + epsilon, hist_CIPSR_bandpass_noise + epsilon))
    KL_divergence_Schuster_bandpass.append(entropy(hist_Schuster_bandpass_sine + epsilon, hist_Schuster_bandpass_noise + epsilon))
    KL_divergence_DL_bandpass.append(entropy(hist_DL_bandpass_sine + epsilon, hist_DL_bandpass_noise + epsilon))

    # now do the same for PAC filtered data
    hist_R_value_PAC_sine, _ = np.histogram(np.array(R2_value_of_PAC_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_R_value_PAC_noise, _ = np.histogram(np.array(R2_value_of_PAC_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_CIPSR_PAC_sine, _ = np.histogram(np.array(CIPSR_of_PAC_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_CIPSR_PAC_noise, _ = np.histogram(np.array(CIPSR_of_PAC_noise_peaks)[indices_noise], bins=value_bins, density=True)
    hist_Schuster_PAC_sine, _ = np.histogram(np.array(Schuster_mag_of_PAC_sin_wave_peak)[indices_sine], bins=value_bins_for_SSmag, density=True)
    hist_Schuster_PAC_noise, _ = np.histogram(np.array(Schuster_mag_of_PAC_noise_peaks)[indices_noise], bins=value_bins_for_SSmag, density=True)
    hist_DL_PAC_sine, _ = np.histogram(np.array(DL_prob_of_PAC_sin_wave_peak)[indices_sine], bins=value_bins, density=True)
    hist_DL_PAC_noise, _ = np.histogram(np.array(DL_prob_of_PAC_noise_peaks)[indices_noise], bins=value_bins, density=True)

    # calculate the KL divergence
    KL_divergence_R_value_PAC.append(entropy(hist_R_value_PAC_sine + epsilon, hist_R_value_PAC_noise + epsilon))
    KL_divergence_CIPSR_PAC.append(entropy(hist_CIPSR_PAC_sine + epsilon, hist_CIPSR_PAC_noise + epsilon))
    KL_divergence_Schuster_PAC.append(entropy(hist_Schuster_PAC_sine + epsilon, hist_Schuster_PAC_noise + epsilon))
    KL_divergence_DL_PAC.append(entropy(hist_DL_PAC_sine + epsilon, hist_DL_PAC_noise + epsilon))
```
