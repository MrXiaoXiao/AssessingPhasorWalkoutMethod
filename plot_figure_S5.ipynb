```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# three models that takes the raw/bandpassed/PAC data as input, respectively
# epoch,accuracy,loss,lr,val_accuracy,val_loss
raw_log = pd.read_csv('phasor_walkout_DL_classification_training_val_raw.log')
bandpassed_log = pd.read_csv('phasor_walkout_DL_classification_training_val_bandpass.log')
PAC_log = pd.read_csv('phasor_walkout_DL_classification_training_val_PAC.log')

# plot the training and validation loss
plt.figure(figsize=(12, 8))
plt.subplot(3, 1, 1)
plt.plot(raw_log['epoch'], raw_log['loss'], label='raw training loss', color='r')
plt.plot(raw_log['epoch'], raw_log['val_loss'], label='raw validation loss', color='b')
min_val_loss_epoch = raw_log['val_loss'].idxmin()
min_val_loss = raw_log['val_loss'].min()
plt.scatter(min_val_loss_epoch, min_val_loss, color='k',marker='x', s=300, zorder=10, label='min val loss', edgecolors='k')

plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.title('Training and Validation Loss (Model for Raw data)', fontsize=16)
plt.xlim([0, 100])
plt.ylim([0.0, 0.4])

plt.subplot(3, 1, 2)
plt.plot(bandpassed_log['epoch'], bandpassed_log['loss'], label='bandpassed training loss', color='r')
plt.plot(bandpassed_log['epoch'], bandpassed_log['val_loss'], label='bandpassed validation loss', color='b')
min_val_loss_epoch = bandpassed_log['val_loss'].idxmin()
min_val_loss = bandpassed_log['val_loss'].min()
plt.scatter(min_val_loss_epoch, min_val_loss, color='k',marker='x', s=300, zorder=10, label='min val loss')

plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.title('Training and Validation Loss (Model for Bandpassed data)', fontsize=16)
plt.xlim([0, 100])
plt.ylim([0.0, 0.4])

plt.subplot(3, 1, 3)
plt.plot(PAC_log['epoch'], PAC_log['loss'], label='PAC training loss', color='r')
plt.plot(PAC_log['epoch'], PAC_log['val_loss'], label='PAC validation loss', color='b')
min_val_loss_epoch = PAC_log['val_loss'].idxmin()
min_val_loss = PAC_log['val_loss'].min()
plt.scatter(min_val_loss_epoch, min_val_loss, color='k',marker='x', s=300, zorder=10, label='min val loss')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.legend(fontsize=14)
plt.title('Training and Validation Loss (Model for PAC data)', fontsize=16)
plt.xlim([0, 100])
plt.ylim([0.0, 0.4])
plt.tight_layout()
plt.show()
plt.close()
```
