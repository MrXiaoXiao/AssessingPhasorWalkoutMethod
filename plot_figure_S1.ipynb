```python
from phw_lib import *
```

```python
# create a constant sequence
fs = 0.5
dt = 1/fs
time_duration = 12000
n_samples = int(time_duration/dt)
t = np.arange(0, time_duration, dt)
x = np.ones(n_samples)*800

sin_wave_freq = 5*1e-3
sin_wave_phase = np.pi/4
sin_wave_amp = 1.0
sin_wave = sin_wave_amp * np.sin(2 * np.pi * sin_wave_freq * t + sin_wave_phase)
```

```python
plt.figure(figsize=(10, 8))
plt.subplot(4, 1, 1)
plt.plot(t, x, 'k')
plt.xlim([0, time_duration])
plt.xlabel('Time (s)', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.title('Constant sequence', fontsize=18)
plt.text(-0.05, 1.10, '(a)', transform=plt.gca().transAxes, size=20, weight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

plt.subplot(4, 1, 2)
visualize_walkout_for_subplot(x, dt, sin_wave_freq)
plt.title('{:.2f} mHz Walkout'.format(sin_wave_freq*1e3), fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.text(-0.05, 1.10, '(b)', transform=plt.gca().transAxes, size=20, weight='bold')

plt.subplot(4, 1, 3)
sine_plus_constant = x + sin_wave
plt.plot(t, sine_plus_constant, 'k')
plt.xlim([0, time_duration])
plt.xlabel('Time (s)', fontsize=16)
plt.ylabel('Amplitude', fontsize=16)
plt.title('Constant + 5.0 mHz Sine', fontsize=18)
plt.text(-0.05, 1.10, '(c)', transform=plt.gca().transAxes, size=20, weight='bold')
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)


plt.subplot(4, 1, 4)
visualize_walkout_for_subplot(sine_plus_constant, dt, sin_wave_freq)
plt.title('{:.2f} mHz Walkout'.format(sin_wave_freq*1e3), fontsize=18)
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)
plt.text(-0.05, 1.10, '(d)', transform=plt.gca().transAxes, size=20, weight='bold')

plt.tight_layout()
plt.savefig('figure_S1.jpg', dpi=300)
plt.show()
plt.close()
```
